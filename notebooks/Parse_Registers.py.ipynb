```python
raw_reg = r'../doc/raw_registers.txt'
```

```python
with open(raw_reg) as reg_file:
    lines = reg_file.readlines()
    for line in lines:
        if line.startswith('0x') and len(line) > 10 and 'Reserved' not in line:
            cols = line.split()
            print(cols[:8],' '.join(cols[8:]))
```
